# Data Cleaning

```python
# drop missing values
df.dropna(inplace=True)
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# find  and drop duplicates value 
print("Total Duplicates Value:",df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
# after drop duplicates
print("Total Duplicates Value:",df.duplicated().sum())
```

```python
# change datatype
df["id"]=df["id"].astype("object")
df["host_id"]=df["host_id"].astype("object")
df.dtypes
```

# Visualisations

# Univariate analysis 

```python
# find outliers
sns.boxplot(data=df,y="price")
plt.title("Find Outliers From Price Column")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# remmove outliers
df=df[df["price"]<1500]
sns.boxplot(data=df,y="price")
plt.title("Find Outliers From Price Column")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="price",bins=100)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# find availability 
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="availability_365")
plt.title("availability_365 Distribution")
plt.xlabel("availability_365")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
df.groupby("neighbourhood_group")["price"].mean()
```

```python
# get price per bed
df["price per bed"]=df["price"]/df["beds"]
df.head()
```

```python
df.groupby("neighbourhood_group")["price per bed"].mean()
```

```python
# price dependency on neighneighborhood 
sns.barplot(data=df,x="neighbourhood_group",y="price",hue="room_type")
plt.title("Price Dependency On Neighneighborhood")
plt.xlabel("Neighbourhood_Group")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=df,vars=["price","minimum_nights","number_of_reviews","availability_365"],hue="room_type")
```

```python
# number of reviews by price
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x="number_of_reviews",y="price",hue="neighbourhood_group")
plt.title("Locality And Review Dependency")
plt.show()
```

```python
# Geographical Distribution Of AirBnb Listing
plt.figure(figsize=(7,6))
sns.scatterplot(data=df,x="longitude",y="latitude",hue="room_type")
plt.title("Geographical Distribution Of AirBnb Listing")
plt.show()
```
